```python
import os
import cv2
import random
import numpy as np
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/test2/MEL_techno&semi/'

train_path='/content/drive/MyDrive/Colab Notebooks/test2/MEL_techno&semi_split/train/'
test_path='/content/drive/MyDrive/Colab Notebooks/test2/MEL_techno&semi_split/test/'

os.listdir(path),len(os.listdir(path))
```

```python
def splitting_array(images_list,split_perentage):
  
  rng = np.random.default_rng()
  rng.shuffle(images_list, axis = 0)

  images_arr = np.array(images_list)

  # randomly permute the indices of the array
  indices = np.random.permutation(len(images_arr))

  # split the indices into two parts
  split_index = int(len(images_arr)*(split_perentage/100))
  if split_index < 2:
    split_index=2 
  indices_1 = indices[:split_index]
  indices_2 = indices[split_index:]

  print("test :",len(indices_1)," | train : ",len(indices_2))
  # divide the original array into two random subsets
  test = images_arr[indices_1]
  train = images_arr[indices_2]
  return (train,test)
```

```python
def saving_test_train_split(image_path,saving_path):
  img = cv2.imread(image_path)
  cv2.imwrite(saving_path, img)
```

```python
split_perentage = 12
train = []
test = []
images_list = []
words = os.listdir(path)
for word in words:

  if not os.path.exists(train_path+word):
    os.mkdir(train_path+word)
  if not os.path.exists(test_path+word):
    os.mkdir(test_path+word)

  images_list = os.listdir(path+word)
  train,test = splitting_array(images_list,split_perentage)
  print(word)
  print(len(test),len(train))
  
  for tr in train:
    saving_test_train_split(path+word+'/'+tr,train_path+word+'/'+tr)
  
  for tes in test:
    saving_test_train_split(path+word+'/'+tes,test_path+word+'/'+tes)
```
